```python
# a program to reverse a string using a while loop
s = "Python"
reversed_s = ""
index = len(s) - 1

while index >= 0:
    reversed_s += s[index]
    index -= 1

print(f"Reversed string: {reversed_s}")
```

```python
#15 to calculate the factorial of a number provided by the user using a while loop.
try:
n = int(input("Enter non-negative integer for factorial: "))
if n < 0:
print("Factorial not defined.")
else:
result = 1
 i = n
while i > 1:
result *= i
i -= 1
print(f"Factorial of {n} is: {result}")
except ValueError:
    print("Invalid input.")
```

#7.What is the difference between “==” and “is” operators in Python.
** "==" (Equality): Checks if the values of the two operands are equal.

>> Example
a = [1, 2]
b = [1, 2]

a is b is False because they are stored in different memory locations.
